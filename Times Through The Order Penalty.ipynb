```python
# Importing packages

import pandas as pd
import numpy as np
```

```python
# Import data file containing first at bats in top and bottom of first inning from 2012-2016
# This data file is used to determine home and away starting pitchers for each game
starting_pitchers = pd.read_csv("Starting Pitchers/SP Play Event.csv")
starting_pitchers.head()
```

```python
# Import data file containing all at bats by pinch hitters from 2012-2016
pinch_hitting = pd.read_csv("Pinch Hitting Results/PH Play Event.csv")
pinch_hitting.head()
```

```python
# Merge pinch_hitting and starting_pitchers data frames to include only pinch hitting at bats against starting pitchers
pinch_hit_against_starter_merged = pd.merge(pinch_hitting, starting_pitchers, on=['Pitcher','Date'], how='inner')

# Retain only the important columns
pinch_hit_against_starter = pinch_hit_against_starter_merged[['Date','Batter_x','Tm_x','Opp_x','Pitcher','Outcome_x','Inn_x','RoB_x','Out_x','WPA_x','RE24_x','Play Description_x']]

# Renaming columns
pinch_hit_against_starter.columns = ['Date','Batter','Tm','Opp','Pitcher','Outcome','Inn','RoB','Out','WPA','RE24','Play Description']

# Quick summary
print "Total number of starting pitchers from 2012 to 2016: ", len(starting_pitchers)
print "Total number of pinch hitting at bats from 2012 to 2016: ", len(pinch_hitting)
print "Total number of pinch hitting at bats against a starting pitcher from 2012 to 2016: ", len(pinch_hit_against_starter)

pinch_hit_against_starter.head()
```

```python
# Import data file with all at bats in first three innings of all games from 2012-2016
# This data file will include all starting batters' first at bats
first_three_inn = pd.read_csv("First Time Through Hitting Results/FTT Hit Event.csv")
first_three_inn.head()
```

```python
# Remove at bats corresponding to second at bat by starting batter
only_first_abs = first_three_inn.drop_duplicates(subset = ['Date','Batter'])

# Remove at bats corresponding to at bats not against starting pitcher
first_time_through_merged = pd.merge(only_first_abs, starting_pitchers, on=['Date','Pitcher'], how='inner')

# Retain only the important columns
first_time_through = first_time_through_merged[['Date','Batter_x','Tm_x','Opp_x','Pitcher','Outcome_x','Inn_x','RoB_x','Out_x','WPA_x','RE24_x','Play Description_x']]

# Renaming columns
first_time_through.columns = ['Date','Batter','Tm','Opp','Pitcher','Outcome','Inn','RoB','Out','WPA','RE24','Play Description']

# Quick summary
print 'There were %d unique first time at bats in innings 1-3 from 2012 to 2016' % len(only_first_abs)
print 'There were %d unique first time at bats in innings 1-3 against starting pitchers from 2012 to 2016' % len(first_time_through)

first_time_through.head()
```

```python
# Figure out pinch_hit_against_reliever, data frame describing how pinch hitters did against relievers
# This amounts to finding the complement of pinch_hit_against_starter
# Accomplish this by taking union of pinch_hitting and pinch_hit_against_starter and then dropping duplicates

# Retain only important columns
pinch_hitting_important = pinch_hitting[['Date','Batter','Tm','Opp','Pitcher','Outcome','Inn','RoB','Out','WPA','RE24','Play Description']]

union = pd.concat([pinch_hitting_important, pinch_hit_against_starter], sort=True)
pinch_hit_against_reliever = union.drop_duplicates(keep=False)

# Reordering columns because pd.concat did a funky re-ordering
cols = ['Date','Batter','Tm','Opp','Pitcher','Outcome','Inn','RoB','Out','WPA','RE24','Play Description']
pinch_hit_against_reliever = pinch_hit_against_reliever[cols]

# Quick summary
print "Total number of pinch hitting at bats against a reliever from 2012 to 2016: ", len(pinch_hit_against_reliever)

pinch_hit_against_reliever.head()
```

```python
## If first ABs by starters were equivalent to pinch hitters against same starter (accounting for pinch hitting penalty), this would seem to imply fatigue not majorly important and perhaps familiarity more important for pitcher decline
## If first ABs by starters were worse than pinch hitters against same starter (acounting for pinch hitting penalty), this would seem to imply fatigue of primary importance
## Need to estimate what pinch hitting penalty should be in RE24 to conclude

print 'The average RE24 of starters first time through the lineup is', first_time_through['RE24'].mean()
print 'The average RE24 of pinch hitters against the same pitchers is', pinch_hit_against_starter['RE24'].mean()
```

```python
## Pinch hitters vs. starters are essentially the same as pinch hitters vs. relievers.
## Since starters are roughly equivalent to relievers with platoon advantage, this seems to imply fatigue not a huge factor

print 'The average RE24 of pinch hitters against starters is', pinch_hit_against_starter['RE24'].mean()
print 'The average RE24 of pinch hitters against relievers is', pinch_hit_against_reliever['RE24'].mean()
```
